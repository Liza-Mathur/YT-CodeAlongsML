```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.info()
```

```python
iris['species'].unique()
```

```python
df = iris[['sepal_length','sepal_width','petal_length','petal_width']]
df
```

```python
species = iris['species'].map({'setosa':1,'virginica':2,'versicolor':3})
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pcs = pca.fit_transform(df)
plt.scatter(pcs[:,0],pcs[:,1], c=species)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
y_means = kmeans.fit_predict(pcs)
plt.scatter(pcs[:,0],pcs[:,1], c=y_means)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++')
y_means_init = kmeans.fit_predict(pcs)
plt.scatter(pcs[:,0],pcs[:,1], c=y_means_init)
```

```python
df2 = df[['sepal_length','petal_length']]
```

```python
df3 = iris[['sepal_length','petal_length','species']]
```

```python
kmeans2 = KMeans(n_clusters=3, init='k-means++')
y_pred_df2 = kmeans2.fit_predict(df2)
plt.scatter(df2['sepal_length'],df2['petal_length'], c=y_pred_df2)
```
