```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.iloc[:,:-1]
```

```python
df['species'].unique()
```

```python
df['species'].value_counts()
```

```python
df = df[df['species']!='setosa']
```

```python
df['species'].unique()
```

```python
df['species'] = df['species'].map({'versicolor' : 1 , 'virginica':0})
```

```python
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=41)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
lgr = LogisticRegression()
param = {'penalty':['l1','l2','elasticnet'], 'C':[1,2,3,4,5,10,20,30],'max_iter':[100,200,300,500]}
```

```python
grid_model = GridSearchCV(lgr, param_grid= param, cv = 7, scoring='accuracy')
```

```python
grid_model.fit(x_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
lr = LogisticRegression(C = 20, max_iter=100, penalty='l2')
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(classification_report(y_pred, y_test))
```

```python
accuracy_score(y_test , y_pred)
```
