```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
df_obj = encoder.fit_transform(df_obj)
```

```python
df_obj = df_obj.drop('RainTomorrow',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_new, y, test_size=0.3, random_state=101)
```

```python
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[1,2,3,5,10], 'penalty':['l1','l2','elasticnet'], 'max_iter' : [100,200,300,500]}
grid_model = GridSearchCV(lr, param_grid= param_grid, scoring='accuracy', cv=5)
```

```python
grid_model.fit(x_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
logModel = LogisticRegression(C=3, max_iter=100, penalty = 'l2')
logModel.fit(x_train, y_train)
y_pred = logModel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
