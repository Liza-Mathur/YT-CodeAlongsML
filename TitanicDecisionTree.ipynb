# Analysis - 
- Pclass 1 passengers were mostly found on decks A, B, C, and D.
- Higher survival rates were observed on higher decks (A, B, C).
- Women had a much higher survival rate than men, even though they were fewer in number. However, on G deck, women's survival rate was extremely low.
- Passengers on C deck had high survival rates, and they were mostly from A, B, or C decks.
- Passengers who embarked from Queenstown (Q) mostly ended up on D deck, and those on D deck also had a good survival rate.
- Passengers from Southampton (S) were mostly on A, B, or C decks, but their survival rate was lower.
- Age vs. Survival across decks:
   - A deck: Equal number of deaths and survivors.
   - B deck: Passengers aged 20-40 mostly survived, while older ones died.
   - C deck: People above 40 died, but those below 40 mostly survived.
   - D deck: Survival was random, regardless of age.
   - E deck: Passengers above 50 mostly died, while those below 50 survived.
   - F deck: Passengers aged 25 or younger survived, but those older than 25 mostly died.
   - G deck: Most passengers were under 30, and survival was mixed (some died, some survived).
   - Only G,F, AND C had kids

```python
import numpy as np

# Define a function to assign the deck based on 'who', 'survived', and 'pclass'
def assign_deck(row):
    if row["who"] == "woman" and row["survived"] == 1:
        if row["pclass"] == 1:
            return "A"
        elif row["pclass"] == 2:
            return "D"
        elif row["pclass"] == 3:
            return "F"
    elif row["who"] == "woman" and row["survived"] == 0:
        if row["pclass"] == 1:
            return "C"
        elif row["pclass"] == 2:
            return "E"
        elif row["pclass"] == 3:
            return "G"
    return np.nan  # Default value if no condition is met

# Apply the function to the 'deck_assigned' column where it's NaN
df["deck"] = df["deck"].fillna(df.apply(assign_deck, axis=1))

# Display the updated DataFrame
df.head()
```
