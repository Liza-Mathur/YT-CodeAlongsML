```python
from sklearn.datasets import fetch_california_housing
```

```python
ds = fetch_california_housing()
ds
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(ds.data , columns = ds.feature_names)
```

```python
y = pd.DataFrame(ds.target, columns = ['MedHouseVal'])
```

```python
df
```

```python
y
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(y)
plt.ylim(4.825, 4.83)
```

```python
outliers = y[y['MedHouseVal'] > 4.825].index
```

```python
len(outliers)
```

```python
outliers
```

```python
df_clean = df.drop(outliers)
y_clean = y.drop(outliers)
```

```python
y_clean.loc[90]
```

```python
df_clean.isnull().sum()
```

```python
y_clean.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_clean, y_clean, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred_lr , y_test)
```

```python
y_pred_lr
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=42)
tree.fit(x_train, y_train)
```

```python
y_pred_dtr = tree.predict(x_test)
```

```python
r2_score(y_test, y_pred_dtr)
```

```python
from xgboost import XGBRegressor
xgboost = XGBRegressor(n_estimators=300, learning_rate=0.1, max_depth=6, random_state=42)
xgboost.fit(x_train, y_train)
```

```python
y_pred_xgb = xgboost.predict(x_test)
```

```python
r2_score(y_test, y_pred_xgb)
```
