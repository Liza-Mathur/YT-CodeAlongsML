```python
from sklearn.datasets import load_diabetes
```

```python
ds = load_diabetes()
```

```python
ds
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(ds.data, columns = ds.feature_names)
y = pd.DataFrame(ds.target , columns = ['target'])
```

```python
df
```

```python
y
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(y['target'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('sex', axis=1), y, test_size=0.15, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_test, y_pred))
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python

# Merge your datasets if they are separate
data = pd.concat([df, y], axis=1)

# Compute correlation with target variable
correlation_matrix = data.corr()

# Display correlation of features with the target
print(correlation_matrix["target"].sort_values(ascending=False))
```

```python
from sklearn.model_selection import cross_val_score  

model = LinearRegression()  
cv_scores = cross_val_score(model, df, y, cv=5, scoring="r2")  

print("Cross-Validation R² Scores:", cv_scores)  
print("Mean R²:", cv_scores.mean())  
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0) 
ridge.fit(x_train, y_train)

y_pred = ridge.predict(x_test)
print(r2_score(y_test, y_pred))
```
